### ***Correlation Heatmap***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

file_path = r"C:\Users\HP\OneDrive\Desktop\AIML_NOTES\dataset\ML470_S3_HR_Data_Practice.xlsx"
df = pd.read_excel(file_path)

le_salary = LabelEncoder()
df['salary_enc'] = le_salary.fit_transform(df['salary'])

le_department = LabelEncoder()
df['Department_enc'] = le_department.fit_transform(df['Department'])

numeric_df = df.select_dtypes(include=['number'])
correlation = numeric_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation,
    annot=True,
    cmap='viridis',
    linewidths=0.5,
    square=True
)
plt.title("Correlation Heatmap - HR Data")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### ***Boxplot***

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = r"C:\Users\HP\OneDrive\Desktop\AIML_NOTES\dataset\ML470_S3_HR_Data_Practice.xlsx"
df = pd.read_excel(file_path)

# Numeric columns
box_cols = [
    'satisfaction_level',
    'last_evaluation',
    'number_project',
    'average_montly_hours',
    'time_spend_company',
    'Work_accident',
    'left',
    'promotion_last_5years'
]

box_df = df[box_cols]

# Boxplot (Matplotlib style  same as image)
plt.figure(figsize=(12, 4))
plt.boxplot(
    box_df.values,
    labels=box_cols,
    patch_artist=True,
    boxprops=dict(facecolor='red')
)

plt.title("Distribution of Numeric HR Features (Boxplots)")
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
